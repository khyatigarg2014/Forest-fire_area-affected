```python
d={'sun':0,'mon':1,'tue':2,'wed':3,'thu':4,'fri':5,'sat':6}
data['day']=data['day'].apply(lambda x: d[x])
m={'jan':0,'feb':1,'mar':2,'apr':3,'may':4,'jun':5,'jul':6,'aug':7,'sep':8,'oct':9,'nov':10,'dec':11}
data['month']=data['month'].apply(lambda x: m[x])
```

# #Modelling

# Loading libraries

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error
```

We have to separate the independent and dependent variables for preparing model decieded to drop 'month','rain' and 'area' .

```python
knn=KNeighborsRegressor(n_neighbors=7,weights='distance')
scaler=StandardScaler()
X=data.drop(['day','rain','month','area'],axis=1)
y=data['area']
```

for checking the accuracy of the model we have to break it in testing and training set but as the data is very less so we just took 20% of the data as test set because training set must have sufficient data to train the model accurately. Also during splitting randomized state is varied.

Also as the area was in huge range so we took its log transformation for modelling and also standardized scaling was done as features were in too much scattered ranges.

```python
train_x,test_x,train_y,test_y=train_test_split(X,y,test_size=0.2,random_state=35)
train_y=np.log(train_y+1)
scaler.fit_transform(train_x)
scaler.transform(test_x)
scale=StandardScaler()
train_y=pd.DataFrame(train_y)
scale.fit_transform(train_y)
```

A number of regression models are used and also their parameters were varied. Definition of these models are defined in sklearn libraries.

```python
lr=LinearRegression(normalize=True)
lr.fit(train_x,train_y)
pred=lr.predict(test_x)
pred=scale.inverse_transform(pred)
pred=np.exp(pred)-1
print(mean_squared_error(test_y,pred))
print(mean_absolute_error(test_y,pred))
```

```python
abr=AdaBoostRegressor(n_estimators=150,loss='square')
abr.fit(train_x,train_y)
pred=abr.predict(test_x)
pred=np.exp(pred)-1
print(mean_squared_error(test_y,pred))
```

```python
gbr=GradientBoostingRegressor(n_estimators=10)
gbr.fit(train_x,train_y)
pred=abr.predict(test_x)
pred=np.exp(pred)-1
print(mean_squared_error(test_y,pred))
```

so we got around same accuracy in  adaboost and gradientboostingregressor so used adaboosting because of its less complexity.
